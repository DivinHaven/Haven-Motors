### Slicing, indexing

Extract elements from an array (similar to the one used for Python lists).

```python
a = np.linspace(10, 20, 11)
b = np.array([[1, 2, 3,], [4, 5, 6]])

## Subsetting
a[2]    # Select the element at the 2nd index
b[1, 2]    # Select the element at row 1 column 2 (equivalent to b[1][2])

## Slicing
a[0:2]   # Select items at index 0 and 1
b[0:2, 1]    # Select items at rows 0 and 1 in column 1

## Boolean Indexing
a[a<2]    # Select elements from a less than 2
```

### Array manipulation

You can also use funtions to reshape an array, add/remove elements, stack or split arrays.

```python
a = np.linspace(0, 10, 11)
b = np.sin(a)

np.concatenate([a, b])    # Concatenante arrays
np.vstack((a, b)).shape    # Stack arrays vertically (row-wise)
# np.c_[a, b]    # idem
np.column_stack((a, b))    # Create stacked column-wise arrays
# np.c_[a, b]    # idem
```

```python
np.append(a ,b)    # Append items to an array
np.insert(a, 1, 5)    # Insert items in an array
c = np.delete(a, 0)    # Delete items from an array

print(a)
c
```

### Basic maths

Usual math operations and functions can be applied to arrays, depending on their dimension.

```python
f = np.array([1, 2, 3, 4])
g = np.array([2, 4, 6, 8])

f + g    # Add
f - g    # Subtract
f * g    # Multiply
f / g    # Divide

f**2    # Squared
np.exp(f)     # Exponentiation
np.sqrt(f)     # Square root
np.sin(f)     # Print sines of an array
np.cos(g)     # Element-wise cosine
np.log(f*g)     # Element-wise natural logarithm
np.abs(g)    # Absolute value
g.dot(f)     # Dot produc


## Round
np.floor([1.2, 2.5])    # down
np.ceil([1.2, 2.5])    # Up


## Comparison
print(f == g)    # Element-wise comparison
print(f < 4)
np.array_equal(f, g)    # Array-wise comparison
```

### Statistics functions

Numpy provides basic statistics functions. For more advanced options, it's recommended to use dedicated librairies (usually based on `numpy`), like `statsmodels`.

```python
a = np.linspace(10, 20, 11)
b = np.array([[1, 2, 3,], [4, 5, 6]])
print(b)

## Min/max
a.min()    # Array-wise minimum value, 
print(np.max(a))
b.min(axis=0)    # Maximum value of an array row
a_i = a.argmax()    # Get index for max value
print(a_i)
max_a = a[a_i]    # Get numbers corresponding to indexes 
print(max_a)

## Sum
a.sum()    # Array-wise sum
b.cumsum(axis=1)    # Cumulative sum of the elements


## Average, standard deviation
a.mean()    # Mean
np.median(b)     # Median
np.corrcoef(b)    # Correlation coefficient
np.std(a)    # Standard deviation
```

### Basic plot

Considering $x$ as the distance in mm, we aim at plotting $y = x^2$. The data set is made of two arrays, `x` and `y`, of same dimension. 

```python
## Import libraries
import numpy as np
import matplotlib.pyplot as plt

## Prepare the data set
x = np.linspace(0, 10, 100)
y = x**2

## Create figure
plt.figure()

## Display plot
plt.plot(x, y)

## Add a title and axis labels
plt.title('Distance squared chart')
plt.xlabel('Distance (mm)')
plt.ylabel('Squared distance')

## Save the plot as png (make sure that test/ folder exists)
plt.savefig('../test/save_plot.png')

## Show plot
# if not in Jupyter Notebook use
# plt.show()
```

### Multiple plots
You can print multiple plots at once. It's arranged along a grid. Define the row and column to print the plot with a number assigned to the plot.


```python
plt.figure()

# First plot
plt.subplot(1, 2, 1)    # 1 line, 2 columns, plot number 1
plt.plot(x, y)

# Second plot ('r' makes the line red)
plt.subplot(1, 2, 2)    # 1 line, 2 columns, plot number 2
plt.plot(y, x, 'r')
```

### Object-oriented figures

So far, the user relied on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting. Here, the user explicitly creates Figures and Axes, and call methods on them (object-oriented style). The figure is an *object* that contains all the plot elements.

It's advantageous for multiple figures or multiple plots in a figure.

```python
## Create the figure

fig = plt.figure(figsize=(5, 4))

# Adds axes with a left, bottom, width and height that ranges from 0 to 1 which is the percent of the canvas you want to use
axes_1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])

# Plot on the axes (If you want a label associated with the legend add it with label)
axes_1.plot(x, y, label='y vs x')

# Make a second plot on the same axis
axes_1.plot(y, x, label='x vs y')


## Set lables and title
axes_1.set_xlabel('Distance')
axes_1.set_ylabel('Distance squared')
axes_1.set_title('Distance squared chart')

# Add an optional legend
axes_1.legend(loc=0)

# Create a second axis inside of others
axes_2 = fig.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.plot(x, y, 'r')
axes_2.set_xlabel('Distance (mm)')
axes_2.set_ylabel('Days squared')
axes_2.set_title('Small d squared chart')
```

### Subplots

You can define multiple plots with subplots and it handles creating the axes objects.

```python
# axes_2 is a list of axes objects
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3)

# You can access the plots by index
axes_2[0].plot(x, x)

axes_2[1].plot(x, y, 'r')
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')

axes_2[2].plot(y, x)

# Put space between axis
plt.tight_layout()
```

### Customizing plot

Many plot properties can be customized: colors, colormap, linestyle, markers, plot limits, axis layout, etc. Text, mathmatical expression and annotations can also be included. The full list of option and various examples are available on the [official Matplotlib documentation](https://matplotlib.org/stable/index.html). Below are some useful tips.

#### Colors
* Default colors (b: blue, g: green, r: red, c: cyan, m: magenta, y: yellow, k: black, w: white)

* Hexcode colors: `color = "#eeefff"`

* Web color names (see a list [here](https://en.wikipedia.org/wiki/Web_colors)): `color = "burlywood"`

* `color = "0.75"` creates a 75% gray

* `alpha` defines the percentage of opacity

#### Line styles
* The default line width is 1. To double it, put it 2.

* Line style can be provided using symbols like `'-'` (plain) or `'--'` (dashed).

#### Markers
* `markerfacecolor` changes the marker fill color

* `markeredgecolor` changes the marker stroke color

* `markeredgewidth` changes the markers stroke size

* Marker style uses symbols like `'o'` or `'d'`

#### Text and annotation
* Use the $\LaTeX$ notation for mathematical symbols and equations: e.g. `label=r'$\lambda^2$'`

#### Limits and layout
* Define the plot range using `xlim` and `ylim`method 

* Use `subplots_adjust` or `tight_layout`methods to ajust the subplot spacings 


```python
fig_3 = plt.figure(figsize=(6, 4))
axes_3 = fig_3.add_axes([0, 0, 1, 1])

axes_3.plot(x, y, color='white', alpha=.75, lw=2, ls='--',
           marker='o', markersize=7, markerfacecolor='y',
           markeredgecolor='y', label=r'$\lambda^2$')

axes_3.legend()

# Set the lower and upper bound of x & y axis
axes_3.set_xlim([0, 3])
axes_3.set_ylim([0, 8])

# Add a grid, color
axes_3.grid(True, color='white', linewidth=0.4, alpha=0.3)
# Set grid background color
axes_3.set_facecolor('black')

# Add text to plot
axes_3.text(0.5, 3, r'Message = $\lambda \int{x^3} dx$', color="white");
```

## Discrete Fourier transform

* Discrete Fourier transform (DFT) separates its input into components that contribute at discrete frequencies
* It has a great number of applications in digital signal processing (e.g., for filtering)
* The discretized input to the transform is customarily referred to as a signal in the *time* domain. The output is called a *spectrum* or *transform* and exists in the *frequency* domain
* DFT can also be used for signal in the *spatial* domain, the transform being in the *spatial frequency* domain
* Check the [Numpy FFT documentation](https://numpy.org/doc/stable/reference/routines.fft.html#) and this [tutorial](https://zestedesavoir.com/tutoriels/3939/jouons-a-implementer-une-transformee-de-fourier-rapide/quelques-rappels-sur-la-transformee-de-fourier-discrete/#fn-1-rCX9lBy29g) (in french) for more details about the DFT and the Fast Fourier Transform (FFT) algorithm that computes it.

```python
import numpy as np
import matplotlib.pyplot as plt

# We consider an example of a signal defined in the time domain

fs = 1000    # Sampling frequency (Hz), number of sample per seconds
N = 1024    # Number of samples
D = N / fs    # Duration (in s)
dt = D / N    # Time interval

t = np.arange(0, D, dt)    # Time array

# Amplitude of a dummy temporal signal (sum of sinus at different frequency)
f1 = 1.
A = 3 * np.sin(2 * np.pi * f1 * t)

f2 = 4
A += np.sin(2 * np.pi * f2 * t)

f3 = 7   
A += 0.5 * np.sin(2 * np.pi * f3 * t)

plt.figure()
plt.plot(t, A, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
```

```python
# Get the DFT sample frequencies
f = np.fft.fftfreq(N, dt)

# Make the FFT of the signal (a complex signal)
A_fft = np.fft.fft(A)

# Make the inverse FFT (to check that there is no loss of information)
A_i = np.fft.ifft(A_fft)


# Create the figure
plt.figure(figsize = (12, 6))

plt.subplot(121)
plt.plot(np.abs(A_fft)**2)    # DFT array arrangement
# plt.plot(f, np.abs(A_fft)**2)   
# plt.plot(np.fft.fftshift(f), np.abs(np.fft.fftshift(A_fft))**2)

# plt.xlim([-10, 10])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power spectrum')

plt.subplot(122)
plt.plot(t, np.real(A_i), 'r--')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
```
